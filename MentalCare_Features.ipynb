# New DataFrame

```python
df['age group'] = df['How old are you(in year)?'].map({'Under 18': 0, '18-30': 1})
df['gender'] = df['What is your gender identity? '].map({'Male': 1, 'Female': 0})
df["university"] = df["Which types of University are you studying in?"].map({"Private": 1, "Public": 2, "others": 3})
df["department"] = df["Name of the Department"].map({"CSE": 2, "ICE": 1})
df["marital status"] = df["Please indicate your current marital status"].map({"Single": 1, "In a Relationship": 2, "Married": 3})
df["employment status"] = df["Employment Status?"].map({"Student": 1, "Part-time": 2, "Full time": 3})
df["sleeping cycle"] = df["Total sleeping time per day(hours)?"].map({"Below 6": 1, "6-8": 2, "Above 8": 3})
df["social media screening"] = df["Social media Usage per day(hours)?"].map({"Below 1": 1, "1-2": 2, "3-4": 3, "4-6": 4, "Above 6": 5})
df["hangout period"] = df["Daily hangout time(hours)?"].map({"Below 1": 1, "1-2": 2, "3-4": 3, "5-6": 4, "Above 6": 5})
df["academic activity"] = df["Daily Academic Activities(Hours)"].map({"Below 1": 1, "1-2": 2, "3-4": 3, "5-6": 4, "Above 6": 5})
df["financial status"] = df["Is your family financially stable?"].map({"Very Stable": 1, "Somewhat Stable": 2, "Neutral": 3, "Somewhat Unstable": 4, "Very Unstable": 5})
df["family stress"] = df["Do you have family Stress?"].map({"Not at All": 1, "Several Days": 2, "More Than Half The Days": 3, "Nearly Every Day": 4})
df["carrier stress"] = df["Do you have carrier stress?"].map({"Not at All": 1, "Several Days": 2, "More Than Half The Days": 3, "Nearly Every Day": 4})
df["nervous level"] = df["Feeling nervous, anxious, or on edge"].map({"Not at All": 1, "Several Days": 2, "More Than Half The Days": 3, "Nearly Every Day": 4})
df["worries level"] = df["Not being able to stop or control worrying"].map({"Not at All": 1, "Several Days": 2, "More Than Half The Days": 3, "Nearly Every Day": 4, "Not at All;Several Days": 1})
df["things level"] = df["Worrying too much about different things"].map({"Not at All": 1, "Several Days": 2, "More Than Half The Days": 3, "Nearly Every Day": 4})
df["trouble relaxing"] = df["Trouble relaxing"].map({"Not at All": 1, "Several Days": 2, "More Than Half The Days": 3, "Nearly Every Day": 4})
df["restless level"] = df["Being so restless that it is hard to sit still"].map({"Not at All": 1, "Several Days": 2, "More Than Half The Days": 3, "Nearly Every Day": 4})
df["irritation level"] = df["Becoming easily annoyed or irritable"].map({"Not at All": 1, "Several Days": 2, "More Than Half The Days": 3, "Nearly Every Day": 4, "Not at All;Several Days": 1})
df["fear level"] = df["Feeling afraid, as if something awful might happen"].map({"Not at All": 1, "Several Days": 2, "More Than Half The Days": 3, "Nearly Every Day": 4})
df["counsel status"] = df["Have you ever went to a psychologist?"].map({"No": 0, "Yes": 1})
df["suicidal experience"] = df["Are you ever experiencing suicidal thoughts?"].map({"No": 0, "Yes": 1})
df["confidence level"] = df["Confidence level?"].map({"No Confidence": 1, "Low": 2, "Medium": 3, "High": 4})
df["study desire"] = df["How would you rate your desire to study this week?"].map({"No desire to Study": 1, "slightly Desire": 2, "Desire to Study": 3, "Very Motivated to study": 4})
df["anxiety status"] = df["Do you feel anxious?"].map({"Yes": 1, "No": 2, "Yes;No": 1})
```

```python
new_df = df[['age group', 'gender', 'university', 'department', 'marital status', 'employment status', 'sleeping cycle',
             'social media screening', 'hangout period', 'academic activity', 'financial status', 'family stress',
             'carrier stress', 'nervous level', 'worries level', 'things level', 'trouble relaxing', 'restless level',
             'irritation level', 'fear level' , 'counsel status', 'suicidal experience', 'confidence level', 'study desire',
             'anxiety status']]
new_df.head(10)
```
